```python
import pandas as pd
import numpy as np
import time
import sqlite3
path= 'D:/Research/NLP/T3/'
```

```python
data = pd.read_csv(filepath_or_buffer=path+'train_triplets.txt', sep='\t', header=None,
names=['user','song','play_count'])
```

```python
data.head(10)
```

```python
#store in dict
song_dict = {}

with open(path+'train_triplets.txt') as f:
    for line_number, line in enumerate(f):
        song = line.split('\t')[1]
        play_count = int(line.split('\t')[2])
        if song in song_dict:
            play_count +=song_dict[song]
        song_dict.update({song:play_count})
```

```python
import itertools  
out = dict(itertools.islice(song_dict.items(), 3))  
print("Dictionary  is : " + str(out)) 
```

```python
# convert into list
song_list = [{'song':k,'play_count':v} for k,v in song_dict.items()]
```

```python
song_count_df = pd.DataFrame(song_list)
song_count_df.head(10)
```

```python
song_count_df = song_count_df.sort_values(by = 'play_count', ascending = False)
song_count_df
```

```python
song_count_df.to_csv(path+'song_count_df.csv', index = False) 
song_count_df = pd.read_csv(path+'song_count_df.csv')
```
